```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
path = Path("../data/processed/telco_churn_processed.csv")
```

```python
telco_data = pd.read_csv(path)
```

```python
telco_data.head().T
```

```python
telco_data.columns
```

```python
telco_data.info()
```

```python
from sklearn.preprocessing import LabelEncoder

enocoded = LabelEncoder()
```

```python
#Target column, label encoded
telco_data['Churn'] = enocoded.fit_transform(telco_data['Churn'])
```

```python
telco_data.nunique()
```

```python
#Binary categorical columns one hot encoding

binary_categorical = ['gender', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']

binary_telco_data_encoded = pd.get_dummies(telco_data, columns=binary_categorical)
```

```python
# one hot encoding to all(multi-class categorical, ordinal category, tenure_group)

for_encoding_columns =['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaymentMethod', 'Contract', 'tenure_group']

encoded_telco_data = pd.get_dummies(binary_telco_data_encoded, columns=for_encoding_columns)
```

```python
encoded_telco_data.head().T
```

```python
encoded_telco_data=encoded_telco_data.astype(int)
```

```python
output_path = Path("../data/encoded_data/encoded_telco_data.csv")
encoded_telco_data.to_csv(output_path, index=False)
```
