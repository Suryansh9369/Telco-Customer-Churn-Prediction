```python
df_copy=df.copy()
```

```python
df_copy.drop(columns=['customerID'], inplace=True)
```

```python
df_copy['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df_copy.info()
```

```python
categorical_cols = df_copy.select_dtypes(include='object').columns
categorical_cols.shape
```

```python
n_col = 2
n_row = (len(categorical_cols) + n_col - 1) // n_col

plt.figure(figsize=(15, 4 * n_row))

for i, col in enumerate(categorical_cols, 1):
    plt.subplot(n_row, n_col, i)
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
df_copy.loc[df_copy['TotalCharges'].isnull()==True]
```

```python
#removing missing value
df_copy.dropna(how='any', inplace=True)
```

```python
#get the max trnure
df_copy['tenure'].max()
```

```python
#group the tenure in the bins of 12 months
labels = [f"{i}-{i+11}" for i in range(1,73, 12)]

df_copy['tenure_group']=pd.cut(df_copy['tenure'], range(0,73,12), labels=labels, include_lowest=True)
```

```python
df_copy['tenure_group'].value_counts()
```
