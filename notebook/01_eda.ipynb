```python
numerical_cols = df.select_dtypes(include=np.number).columns

plt.figure(figsize=(15, 10))

for i, col in enumerate(numerical_cols):
    plt.subplot(4, 4, i + 1)   # Correct way to create subplots
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))

for i, cols in enumerate(numerical_cols):
    plt.subplot(4, 4, i + 1)
    sns.boxplot(x=df[cols])
    plt.title(col)

plt.tight_layout()
plt.show()
    
```

```python
plt.figure(figsize=(15,10))

for i, cols in enumerate(numerical_cols):
    plt.subplot(4, 4, i + 1)
    sns.scatterplot(x=df[cols])
    plt.title(col)

plt.tight_layout()
plt.show()
```
